```python
import sys
import heapq
import io

input_data = """5 6
1
5 1 1
1 2 2
1 3 3
2 3 4
2 4 5
3 4 6
"""

# sys.stdin을 문자열 버퍼로 교체 (입력 리다이렉션)
sys.stdin = io.StringIO(input_data)
input = sys.stdin.readline

INF = float('inf')

def solve():
    V, E = map(int, input().split())
    print(V, E)
    start_node = int(input())
    print(start_node)

    graph = [[] for _ in range(V + 1)]
    print(graph)

    for _ in range(E):
        u, v, w = map(int, input().split())
        print(u, v, w)
        graph[u].append((v, w))
    print('최종')
    print(graph)
    print('<distance 시작>')

    distance = [INF] * (V + 1)
    print(distance)

    def dijkstra(start):
        q = []
        heapq.heappush(q, (0, start))
        distance[start] = 0

        while q:
            dist, now = heapq.heappop(q)

            if distance[now] < dist:
                continue

            for next_node, weight in graph[now]:
                cost = dist + weight

                if cost < distance[next_node]:
                    distance[next_node] = cost
                    heapq.heappush(q, (cost, next_node))

    dijkstra(start_node)

    # 결과 출력
    print(f"--- 시작점 {start_node}번 노드 기준 최단 경로 ---")
    for i in range(1, V + 1):
        if distance[i] == INF:
            print(f"{i}번 정점 : INF (도달 불가)")
        else:
            print(f"{i}번 정점 : {distance[i]}")

if __name__ == "__main__":
    solve()
```

위 코드를 실행하면 `input_data` 변수에 저장된 문자열이 `sys.stdin.readline()`의 입력으로 사용됩니다. `solve()` 함수는 이 가상 입력으로부터 `V`, `E`, `start_node`, 그리고 각 간선 정보를 읽게 됩니다.

`input_data` 안의 내용은 다음과 같습니다.
1.  **첫 번째 줄**: `V` (정점의 개수) 와 `E` (간선의 개수) 를 공백으로 구분하여 입력합니다. (예: `5 6`)
2.  **두 번째 줄**: `start_node` (시작 정점) 를 입력합니다. (예: `1`)
3.  **이후 `E`개의 줄**: 각 줄마다 `u` (시작 정점), `v` (도착 정점), `w` (가중치) 를 공백으로 구분하여 입력합니다. (예: `1 2 2`)
